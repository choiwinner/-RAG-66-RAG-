```python
#https://levelup.gitconnected.com/stock-market-sentiment-prediction-with-openai-and-python-ea2e2261fcbf
import re
import requests
import pandas as pd
import config as cfg
from eodhd import APIClient # EOD Historical Data로 data API 제공
from langchain.chains import LLMChain
from langchain.prompts import PromptTemplate
from langchain.chat_models import ChatOpenAI
```

```python
api_key = ''
api = APIClient(api_key)
```

```python
resp = api.financial_news(s = "MSFT", from_date = '2024-01-01', to_date = '2024-03-20', limit = 100)
df = pd.DataFrame(resp) # json 출력을 datframe으로 변환
df.tail()
```

```python
#텍스트 데이터 전처리
def clean_text(text):
    cleaned_text = re.sub(r'\s+', ' ', text)
    return cleaned_text.strip()

# 전체 열에 전처리 함수 적용
df['content'] = df['content'].apply(clean_text)
```

```python
#ChatOpenAI 객체 호출할때 파라미터 지정
llm = ChatOpenAI(model = "gpt-4",
                 openai_api_key = 'sk-', 
                 temperature = 0)
```

```python
template = """
Identify the sentiment towards the MSFT stocks from the news article , where the sentiment score should be from -10 to +10 where -10 being the most negative and +10 being the most positve , and 0 being neutral
Also give the proper explanation for your answers and how would it effect the prices of different stocks
Article : {statement}
"""

#Langchain PromptTemplate 기능을 사용하여 프롬프트 생성
prompt = PromptTemplate(template = template, input_variables = ["statement"])
llm_chain = LLMChain(prompt = prompt, llm = llm)
```

```python
print(llm_chain.run(df['content'][13]))
```

```python
#토큰 개수를 계산하는 함수
def count_tokens(text):
    tokens = text.split()  
    return len(tokens)
```

```python
# DataFrame 열에 토큰화 함수 적용
df['TokenCount'] = df['content'].apply(count_tokens)
```

```python
# 토큰 수 임계값 정의(예: 토큰이 2개 이상인 행 유지)
token_count_threshold = 3500

# 토큰 수를 기준으로 필터링하여 새 DataFrame을 만듭니다.
new_df = df[df['TokenCount'] < token_count_threshold]

# 필요하지 않은 경우 새 DataFrame에서 'TokenCount' 열을 삭제
new_df = new_df.drop('TokenCount', axis = 1)

# 인덱스 재설정
new_df = new_df.reset_index(drop = True)
```

```python
template_2 = """
Identify the sentiment towards the MSFT stocks of the news article from -10 to +10 where -10 being the most negative and +10 being the most positve , and 0 being neutral

GIVE ANSWER IN ONLY ONE WORD AND THAT SHOULD BE THE SCORE

Article : {statement}
"""

#Langchain PromptTemplate 기능을 사용하여 프롬프트 생성
prompt_2 = PromptTemplate(template = template_2, input_variables = ["statement"])
```

```python
llm_chain_2 = LLMChain(prompt = prompt_2, llm = llm)
print(new_df['content'][2])
print('')
print('News sentiment: ', llm_chain_2.run(new_df['content'][2]))
```
